# Sistema de Inversiones

```python
#@autor:arizsil #Inversion.py
class Inversion :
    #este es el metodo constructor, que
    #inicializa a los atributos de una clase
    def __init__(self, valor1, valor2, valor3, valor4) :
        self.capital = valor1       #dinero inicial invertido
        self.interes = valor2       #tasa de interes
        self.tiempo = valor3        #tiempo de inversion
        self.nombre = valor4        #nombre de la inversion

#recibe una cantidad y la suma al capital actual
    def aumentar(self, cantidad) :
        self.capital = self.capital + cantidad

#resta una cantidad al capital actual
    def disminuir(self, cantidad2) :
        self.capital = self.capital - cantidad2

#metodo que calcula el monto final de la inversion
    def calcular(self) :
        self.capital = self.capital * self.interes * self.tiempo

    def mostrar(self) :
        print(self.nombre)
```
